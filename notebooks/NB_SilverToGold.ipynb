```python
silver_path="abfss://lakehouse@anniedatalake01.dfs.core.windows.net/silver/customers_clean"
df_silver=spark.read.format("delta").load(silver_path)
##df_silver.write.format("delta").mode("overwrite").save(silver_path)
df_silver.show(5)
```

```python
from pyspark.sql.functions import col,round

df_gold=df_silver.withColumn("BMI", round(col("weight_lb")/(col("height_in")*col("height_in"))*703,2))

df_gold.show(5)
```

```python
gold_path="abfss://synapsedata@anniedatalake01.dfs.core.windows.net/gold/customers_model/"
df_gold.write.format("delta").mode("overwrite").save(gold_path)
```
